```python
import pandas as pd
import matplotlib.pyplot as plt
import os

def generate_charts():
    output_dir = "results/charts/"
    os.makedirs(output_dir, exist_ok=True)

    # 1. Gráficos do Experimento 1 (Escalabilidade)
    df_exp1 = pd.read_csv("results/benchmark_exp1_scalability.csv")
    
    # Filtra por tipo de grafo
    sparse_data = df_exp1[df_exp1["Type"] == "sparse"].groupby("N").mean().reset_index()
    dense_data = df_exp1[df_exp1["Type"] == "dense"].groupby("N").mean().reset_index()

    plt.figure(figsize=(10, 6))
    plt.plot(sparse_data["N"], sparse_data["Dijkstra Time (s)"], marker='o', label="Dijkstra (Esparso)")
    plt.plot(sparse_data["N"], sparse_data["Bellman-Ford Time (s)"], marker='s', label="Bellman-Ford (Esparso)")
    plt.plot(dense_data["N"], dense_data["Dijkstra Time (s)"], marker='^', linestyle='--', label="Dijkstra (Denso)")
    plt.plot(dense_data["N"], dense_data["Bellman-Ford Time (s)"], marker='d', linestyle='--', label="Bellman-Ford (Denso)")
    
    plt.title("Escalabilidade: Dijkstra vs Bellman-Ford")
    plt.xlabel("Número de Vértices (N)")
    plt.ylabel("Tempo de Execução (s)")
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.savefig(f"{output_dir}/escalabilidade.png")
    plt.close()

    # 2. Gráficos do Experimento 2 (Sensibilidade)
    df_exp2 = pd.read_csv("results/benchmark_exp2_sensitivity.csv")

    # Gráfico: k x |P| (Esperado: decrescimento)
    plt.figure(figsize=(8, 5))
    plt.plot(df_exp2["K"], df_exp2["|P|"], marker='o', color='purple')
    plt.title("Sensibilidade do Parâmetro k vs Quantidade de Pivôs (|P|)")
    plt.xlabel("Parâmetro de Profundidade (k)")
    plt.ylabel("Número de Pivôs Encontrados (|P|)")
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.savefig(f"{output_dir}/k_vs_p.png")
    plt.close()

    # Gráfico: |S| x |P|
    plt.figure(figsize=(8, 5))
    plt.scatter(df_exp2["|S|"], df_exp2["|P|"], color='red', s=100, alpha=0.6)
    plt.title("Fronteira (|S|) vs Pivôs (|P|)")
    plt.xlabel("Tamanho da Fronteira (|S|)")
    plt.ylabel("Número de Pivôs (|P|)")
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.savefig(f"{output_dir}/s_vs_p.png")
    plt.close()

    # Histograma: distribuição de |P| / (|W|/k)
    # Proteção contra divisão por zero
    w_over_k = df_exp2["|W|"] / df_exp2["K"]
    razao = df_exp2["|P|"] / w_over_k.replace(0, 1)

    plt.figure(figsize=(8, 5))
    plt.hist(razao, bins=10, color='teal', edgecolor='black', alpha=0.7)
    plt.title("Distribuição da Razão $|P| / (|W|/k)$")
    plt.xlabel("Valor da Razão")
    plt.ylabel("Frequência")
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.savefig(f"{output_dir}/histograma_razao.png")
    plt.close()

    print(f"Gráficos gerados com sucesso na pasta: {output_dir}")

if __name__ == "__main__":
    generate_charts()
```
